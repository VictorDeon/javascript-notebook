## Eventos
***

Evento é uma ocorrência interativa que ocorre por intervenção do usuário ou não. Por exemplo: Seguir um link, submeter um formulário, carregar uma página ou recurso do servidor, digitar um texto, mover o mouse e etc.

* **Event Handlers**: São funções que são executadas quando um evento ocorre.

* **Disparador de Eventos**: É o elemento que disparou o evento. Ou seja, o elemento ao qual foi atrelado o event handler.

* **Delegação de eventos**: É especialmente útil quando temos muitos filhos com eventos atrelados individualmente, ex: eventos em todas as tags `<li>` de uma lista ou `<td>` de uma tabela. Delegar um evento é o ato de atrelar um evento ao elemento pai, no caso no `<ul>` ou no `<table>`, economizando recursos da máquina.


```js
const container = document.getElementById("container");
container.addEventListener("click", event => {
    event.stopPropagation();
    if (event.target.nodeName === "BUTTON") {
        console.log("CLICOU NO BOTÃO DENTRO DO CONTAINER!!!!");
    }
    console.log("Porém o evento está no container2: ", event.currentTarget.nodeName);
})
```

Temos um terceiro parâmetro do `addEventListener` que diz se vai ser executado/propagado na fase de bolha (default) ou fase de captura, isso permite modificar a ordem de execução dos eventos da DOM. Se tiver algum evento por exemplo, no document e em um botão ambos vão ser executado ao clicar no botão, o que muda é a ordem, para que seja executado o evento só do botão, utilize o comando `event.stopPropagation()` no handler do botão.

**Manipulando eventos**:

|Método|Descrição|
|------|---------|
|element.addEventListener(evento, função)|Adiciona um evento a um elemento.|
|element.removeEventListener(evento, função)|Remove um evento de um elemento.|

**Eventos**:

|Método|Descrição|
|------|---------|
|click|Insere uma função de evento no clique do botão|
|blur|O evento ocorre quando o elemento perde o foco|
|focus|O evento ocorre quando o elemento recebe foco|
|change|O evento ocorre quando o usuário modifica um elemento, por exemplo, um select.|
|dblclick|O evento ocorre quando ocorrer um duplo clique.|
|keydown|O evento ocorre quando o usuário está pressionando um digito, pega os valores antigos.|
|keypress|O evento ocorre quando o usuário pressiona um digito, pega os valores antigos, só dispara em caracteres alphanumericos.|
|keyup|O evento ocorre quando o usuário tira o dedo do teclado, pega os valores atualizados do form.|
|mousedown|O evento ocorre quando o usuário pressiona o clique no elemento|
|mouseenter|O evento ocorre quando o mouse passa pelo elemento.|
|mouseleave|O evento ocorre quando o mouse sai do elemento.|
|mousemove|O evento ocorre quando o mouse está sobre o elemento.|
|scroll|O evento ocorre quando o usuário da scroll no elemento.|
|submit|O evento ocorre quando o formulário é submetido|
|reset|O evento ocorre quando o input é resetado.|

**Objeto de evento**: Seja com DOM 0 ou DOM 2, os handlers recebem por parâmetro um objeto do tipo evento. Para ter acesso à esse objeto, basta dar um nome ao parâmetro. 

|Método|Descrição|
|------|---------|
|event.currentTarget|Retorna o elemento na qual foi atrelado o handler, ou seja, não necessariamente é o elemento que disparou o evento, pode ser um evento propagado por outro evento.|
|event.target|Retorna exatamente o elemento cujo o evento foi disparado.|
|event.charCode|Retorna o código de caractere Unicode da tecla que acionou o evento onkeypress|
|event.data|Retorna os caracteres inseridos|
|event.keyCode|Retorna o código de caractere Unicode da tecla que acionou o evento onkeypress ou o código da chave Unicode da tecla que acionou o evento onkeydown ou onkeyup|
|event.stopPropagation()|Para a propagação do evento, ou seja, não permite que outros eventos sejam disparados após esse de forma automatica.|
|event.preventDefault()|Cancela o evento se for cancelável, o que significa que a ação padrão que pertence ao evento não ocorrerá|
|event.clientX|Retorna a coordenada horizontal do ponteiro do mouse, relativa à janela atual, quando o evento do mouse foi acionado|
|event.clientY|Retorna a coordenada vertical do ponteiro do mouse, relativa à janela atual, quando o evento do mouse foi acionado|
|event.buttons|Retorna quais botões do mouse foram pressionados quando o evento do mouse foi acionado|
|event.offsetX|Retorna a coordenada horizontal do ponteiro do mouse em relação à posição da borda do elemento alvo|
|event.offsetY|Retorna a coordenada vertical do ponteiro do mouse em relação à posição da borda do elemento alvo|
|event.pageX|Retorna a coordenada horizontal do ponteiro do mouse, relativa ao documento, quando o evento do mouse foi acionado|
|event.pageY|Retorna a coordenada vertical do ponteiro do mouse, relativa ao documento, quando o evento do mouse foi acionado|
|event.screenX|Retorna a coordenada horizontal do ponteiro do mouse, relativa à tela, quando um evento foi acionado|
|event.screenY|Retorna a coordenada vertical do ponteiro do mouse, relativa à tela, quando um evento foi acionado|
