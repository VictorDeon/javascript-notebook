## Memory Leaks
***

Vamos fazer um estouro de memória, ou seja, vamos inserir infinitamente dados no memory heap.

```javascript
let array = [];
for (let i = 5; i > 1; i++) {
    console.log(`Loop infinito ${i}`);
    array.push(i-1);
}
```

### Memory leaks comuns
***

1 - Muitas varíaveis globais, dependendo da quantidade pode estourar a memória.

```javascript
var a = 1;
var b = 2;
var c = 3;
```

2 - Não remover os Event Listeners, por exemplo em um single page application você entrar e sair de uma página muitas vezes sem remover os event listeners dela vai só acumulando e estoura a memória.

```javascript
var element = document.getElementById('button');
element.addEventLister('click', () => console.log("ENTREI"))
```

3 - Uso do setInterval sem para-lo. Esse roda infinitamente se não for parado.

```javascript
setIntervalo(() => console.log("rodando!!!"))
```
